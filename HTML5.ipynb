```python
import os
from jinja2 import Template

# =====================================================================
# 📋 1. 數據契約層：用 Python 定義遊戲世界與核心角色屬性
# =====================================================================
game_config = {
    "title": "Python自動工廠生成的HTML5遊戲",
    "canvas_width": 800,
    "canvas_height": 500,
    "canvas_bg": "#111122",
    "player": {
        "x": 375,
        "y": 400,
        "width": 60,
        "height": 20,
        "speed": 8,
        "color": "#ff0077"
    },
    # 自動生成的敵方陣列數據
    "enemies": [
        {"x": 100, "y": 50, "width": 50, "height": 30, "color": "#00ffcc", "score": 10},
        {"x": 250, "y": 50, "width": 50, "height": 30, "color": "#00ffcc", "score": 10},
        {"x": 400, "y": 50, "width": 50, "height": 30, "color": "#00ffcc", "score": 10},
        {"x": 550, "y": 50, "width": 50, "height": 30, "color": "#00ffcc", "score": 10},
        {"x": 200, "y": 120, "width": 50, "height": 30, "color": "#ffff00", "score": 20},
        {"x": 350, "y": 120, "width": 50, "height": 30, "color": "#ffff00", "score": 20},
        {"x": 500, "y": 120, "width": 50, "height": 30, "color": "#ffff00", "score": 20},
    ]
}

# =====================================================================
# 📐 2. 核心模板層：將 HTML5/JS 結構化代碼封裝為 Jinja2 幾何鋼模
# =====================================================================
html_template_string = """<!DOCTYPE html>
<html lang="zh-TW">
<head>
    <meta charset="UTF-8">
    <title>{{ title }}</title>
    <style>
        body {
            margin: 0;
            background-color: #222;
            display: flex;
            flex-direction: column;
            justify-content: center;
            align-items: center;
            height: 100vh;
            color: #fff;
            font-family: Arial, sans-serif;
        }
        canvas {
            border: 4px solid #fff;
            background-color: {{ canvas_bg }};
            box-shadow: 0 0 20px rgba(0,0,0,0.8);
        }
        .info { margin-bottom: 15px; font-size: 18px; }
    </style>
</head>
<body>

    <div class="info">🤖 這是一個由 Python 後台編譯器自動渲染生成的 HTML5 生產線遊戲</div>
    <canvas id="gameCanvas" width="{{ canvas_width }}" height="{{ canvas_height }}"></canvas>

    <script>
        const canvas = document.getElementById("gameCanvas");
        const ctx = canvas.getContext("2d");

        // 🧠 數據熔斷與注入：由 Python 變數動態渲染而來的 JavaScript 物件
        const player = {
            x: {{ player.x }},
            y: {{ player.y }},
            width: {{ player.width }},
            height: {{ player.height }},
            speed: {{ player.speed }},
            color: "{{ player.color }}"
        };

        // 📥 Python 迴圈展開：自動生成敵人陣列
        const enemies = [
            {% for enemy in enemies %}
            {
                x: {{ enemy.x }},
                y: {{ enemy.y }},
                width: {{ enemy.width }},
                height: {{ enemy.height }},
                color: "{{ enemy.color }}",
                score: {{ enemy.score }},
                alive: true
            },
            {% endfor %}
        ];

        // 鍵盤狀態追蹤
        const keys = { Left: false, Right: false };

        window.addEventListener("keydown", (e) => {
            if (e.key === "ArrowLeft" || e.key === "a") keys.Left = true;
            if (e.key === "ArrowRight" || e.key === "d") keys.Right = true;
        });

        window.addEventListener("keyup", (e) => {
            if (e.key === "ArrowLeft" || e.key === "a") keys.Left = false;
            if (e.key === "ArrowRight" || e.key === "d") keys.Right = false;
        });

        // ⚙️ 運行期邏輯更新 (Update)
        function update() {
            if (keys.Left) player.x -= player.speed;
            if (keys.Right) player.x += player.speed;

            // 邊界物理熔斷
            if (player.x < 0) player.x = 0;
            if (player.x > canvas.width - player.width) player.x = canvas.width - player.width;
        }

        // ⚙️ 運行期畫面渲染 (Draw)
        function draw() {
            // 清空畫布
            ctx.clearRect(0, 0, canvas.width, canvas.height);

            // 畫出玩家 (Python 規格定義)
            ctx.fillStyle = player.color;
            ctx.fillRect(player.x, player.y, player.width, player.height);

            // 畫出活著的敵人
            enemies.forEach(enemy => {
                if (enemy.alive) {
                    ctx.fillStyle = enemy.color;
                    ctx.fillRect(enemy.x, enemy.y, enemy.width, enemy.height);
                }
            });

            // 提示字型
            ctx.fillStyle = "#fff";
            ctx.font = "16px sans-serif";
            ctx.fillText("🎮 控制鍵：使用 ◄ 與 ► 鍵移動底條", 20, 30);
            ctx.fillText("🎯 當前關卡敵人總數: " + enemies.length, 20, 50);
        }

        // 🔄 2D 中樞神經流水線迴圈 (Game Loop)
        function gameLoop() {
            update();
            draw();
            requestAnimationFrame(gameLoop);
        }

        // 🚀 通電，啟動遊戲！
        gameLoop();
    </script>
</body>
</html>
"""

# =====================================================================
# 🚀 3. 執行編譯器：加載數據，渲染並實體導出檔案
# =====================================================================
def generate_html5_game():
    print("[⚙️ Python 遊戲生成中樞啟動...]")

    # 解析鋼模
    template = Template(html_template_string)

    # 進行數據與模板的物理融合
    rendered_html = template.render(game_config)

    # 指定輸出的實體 HTML 檔案路徑
    output_filename = "auto_game.html"

    with open(output_filename, "w", encoding="utf-8") as f:
        f.write(rendered_html)

    print(f"✨ [編譯成功] 已成功自動生成 HTML5 遊戲檔案！")
    print(f"📂 實體檔案路徑 -> {os.path.abspath(output_filename)}")
    print(f"💡 提示：直接雙擊打開該 HTML 檔案即可在本機端瀏覽器遊玩。")

if __name__ == "__main__":
    generate_html5_game()
```
